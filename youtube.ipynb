```python
from googleapiclient.discovery import build
import pandas as pd


API_KEY = ""
YOUTUBE_API_SERVICE_NAME = "youtube"
YOUTUBE_API_VERSION = "v3"


PLAYLIST_ID = "PLerJdmU4J6ajWM3hIAMXzBsnOI9yqe_0s"


youtube = build(YOUTUBE_API_SERVICE_NAME, YOUTUBE_API_VERSION, developerKey=API_KEY)


def get_playlist_videos(playlist_id):
    videos = []
    next_page_token = None

    while True:
        pl_request = youtube.playlistItems().list(
            part="snippet",
            playlistId=playlist_id,
            maxResults=50,
            pageToken=next_page_token
        )

        pl_response = pl_request.execute()

        for item in pl_response['items']:
           title = item['snippet']['title']
           video_id = item['snippet']['resourceId']['videoId']
           video_url = f"https://www.youtube.com/watch?v={video_id}"


           thumbnails = item['snippet'].get('thumbnails', {})
           thumbnail_url = thumbnails.get('high', {}).get('url') or \
                    thumbnails.get('medium', {}).get('url') or \
                    thumbnails.get('default', {}).get('url') or \
                    'No thumbnail available'

           videos.append({
        'title': title,
        'image_url': video_url,
        'video_link': thumbnail_url
         })


        next_page_token = pl_response.get('nextPageToken')

        if not next_page_token:
            break

    return videos


```

```python
import json
video_data = get_playlist_videos(PLAYLIST_ID)


for i, video in enumerate(video_data, start=25):
    video['id'] = i


json_file_path = 'youtube_playlist_data.json'
with open(json_file_path, 'w', encoding='utf-8') as f:
    json.dump(video_data, f, indent=2, ensure_ascii=False)


```
